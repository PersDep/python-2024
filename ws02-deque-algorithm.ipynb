Орки Вороновских гор очень любят ходить к своим шаманам. Так как орков много, к шаманам часто образуются очень длинные очереди. А поскольку много орков в одном месте быстро образуют потасовку, которая мешает шаманам проводить сложные медицинские манипуляции, последние решили установить некоторые правила касательно порядка в очереди.

Обычные орки при посещении шаманов должны вставать в конец очереди. Орки-командиры, знающие особый пароль, встают ровно в ее середину, причем при нечетной длине очереди они встают сразу за центром.

Так как орки также широко известны своим почтительным отношением ко всяческим правилам и законам, шаманы попросили вас написать программу, которая бы отслеживала порядок орков в очереди.

### Формат ввода

В первой строке входных данный записано число N (1 <= N <= 10^5) - количество запросов к программе. Следующие N строк содержат описание запросов в формате:

- `"+ i"` - Орк с номером i (1 <= i <= N) встает в конец очереди.

- `"* i"` - Орк-командир с номером i встает в середину очереди.

- `"-"` - первый орк из очереди уходит к шаманам. Гарантируется, что на момент такого запроса очередь не пуста.

### Формат вывода

Выведите орков в том порядке, в котором они заходят к шаману.

```python
from collections import deque

n = int(input())
queue = deque()
extra = deque()
orcs_in_the_real_queue = 0
orcs_in_the_extra_queue = 0

for _ in range(n):
  query = input().split()
  if query[0] == '+':
    print('Adding an orc to the end: ', query[1])
    extra.append(query[1])
    print('Extra queue status: ', extra)
    print('Real queue status: ', queue)
    orcs_in_the_extra_queue += 1
    print('orcs_in_the_extra_queue on adding: ', orcs_in_the_extra_queue)
    print('orcs_in_the_real_queue on adding: ', orcs_in_the_real_queue)
  elif query[0] == '*':
    print('Adding an orc to the middle: ', query[1])
    extra.appendleft(query[1])
    print('Extra queue status: ', extra)
    print('Real queue status: ', queue)
    orcs_in_the_extra_queue += 1
    print('orcs_in_the_extra_queue on adding: ', orcs_in_the_extra_queue)
    print('orcs_in_the_real_queue on adding: ', orcs_in_the_real_queue)
  else: # removing from the queue
    # print(queue.popleft())
    print('Deleting from the real queue: ', queue.popleft())
    print('Extra queue status: ', extra)
    print('Real queue status: ', queue)
    orcs_in_the_real_queue -= 1
    print('orcs_in_the_extra_queue on removing: ', orcs_in_the_extra_queue)
    print('orcs_in_the_real_queue on removing: ', orcs_in_the_real_queue)

  if orcs_in_the_real_queue < orcs_in_the_extra_queue: # засчёт "равенства" очередей, середина как раз где-то между ними
    queue.append(extra.popleft())
    print('Updating the real queue: ', queue)
    print('Status of the extra queue: ', extra)
    orcs_in_the_real_queue += 1
    orcs_in_the_extra_queue -= 1
    print('orcs_in_the_extra_queue on update of the real queue: ', orcs_in_the_extra_queue)
    print('orcs_in_the_real_queue on update of the real queue: ', orcs_in_the_real_queue)
```
